```python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib
from GeoIllu import GeoIlluLib as gil
```

```python
illu = gil.LatticeCreation()
```

# Normal Square Lattice

```python
pair_x, pair_y, dual_lattice_x, dual_lattice_y, num_pair = illu.SquareLatticeEdge(width=8, 
                                                                                  height=8, 
                                                                                  latconst=1, 
                                                                                  vertical_length=1)

plt.figure()
for i in range(num_pair):
    plt.plot(pair_x[i,:], pair_y[i,:], 'b-', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.gca().set_axis_off()
plt.tight_layout()
plt.savefig('example.svg', format='svg')
```

# Normal Honeycomb

```python
pair_x, pair_y, dual_lattice_x, dual_lattice_y, num_pair = illu.HoneycombLatticeEdge(width=10
                                                                                     , height=int(21), latconst=1, vertical_length=1)

plt.figure()
for i in range(num_pair):
    plt.plot(pair_x[i,:], pair_y[i,:], 'b-', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.gca().set_axis_off()
plt.tight_layout()
plt.savefig('example.svg', format='svg')
```

# Moiré Pattern

```python
pair_x, pair_y, dual_lattice_x, dual_lattice_y, num_pair = illu.HoneycombLatticeEdge(width=21, height=42, latconst=1, rotate=-0.06)
pair_x_2, pair_y_2, leng_2 = illu.HoneycombLatticeEdge(width=21, height=43, latconst=1, rotate=0.06)

plt.figure()
for i in range(num_pair):
    plt.plot(pair_x[i,:], pair_y[i,:], 'r-', linewidth=0.7)
for i in range(leng_2):
    plt.plot(pair_x_2[i,:], pair_y_2[i,:], 'b-', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.gca().set_axis_off()
plt.tight_layout()
plt.savefig('moire.svg', format='svg')
```

# Centered Honeycomb Lattice

```python
pair_x, pair_y, leng = illu.HoneycombLatticeEdge_Center(layer=12, latconst=1, rotate=-0.06)
pair_x_2, pair_y_2, leng_2 = illu.HoneycombLatticeEdge_Center(layer=12, latconst=1, rotate=0.06)

plt.figure()
for i in range(leng):
    plt.plot(pair_x[i,:], pair_y[i,:], 'r-', linewidth=0.7)
for i in range(leng_2):
    plt.plot(pair_x_2[i,:], pair_y_2[i,:], 'b-', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.gca().set_axis_off()
plt.tight_layout()
plt.savefig('moire-center.svg', format='svg')
```

# Customize

```python
pair_x, pair_y, leng = illu.HoneycombLatticeEdge_Center(layer=1, latconst=1)

plt.figure()
for i in range(leng):
    plt.plot(pair_x[i,:], pair_y[i,:], 'k-', linewidth=1.5)

tri_up_x = [-np.sqrt(3)/2,np.sqrt(3)/2,0,-np.sqrt(3)/2]
tri_up_y = [0.5, 0.5,1,0.5]
plt.plot(tri_up_x, tri_up_y,'b--', linewidth=2)
tri_down_x = [-np.sqrt(3)/2,np.sqrt(3)/2,0,-np.sqrt(3)/2]
tri_down_y = [-0.5, -0.5,-1,-0.5]
plt.plot(tri_down_x, tri_down_y,'r--', linewidth=2)
plt.gca().set_aspect('equal', adjustable='box')
plt.gca().set_axis_off()
plt.tight_layout()
plt.savefig('example.svg', format='svg')
```
