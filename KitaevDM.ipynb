## Computing two-points correlation functions
toric code limit

```python
# dimerized in the z-direction
DM.Jx = 0.01
DM.Jy = 0.01
DM.Jz = 1
print(DM.corr(na=5, nb=2))
print(DM.corr(na=5, nb=6))

# dimerized in the x-direction
DM.Jx = 1
DM.Jy = 0.01
DM.Jz = 0.01
print(DM.corr(na=5, nb=2))
print(DM.corr(na=5, nb=6))
```

Wire limit: The inter-wire correlation is always finite 

```python
DM.Jx = 1
DM.Jy = 0.5
DM.Jz = 0.3
print(DM.corr(na=5, nb=2))
print(DM.corr(na=5, nb=6))
```
